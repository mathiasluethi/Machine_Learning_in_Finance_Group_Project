```python
import pandas as pd
import numpy as np

# Load in data from the Wharton Database
df = pd.read_csv('Data/WhartonData.csv', sep=',', header=0)

df = df.drop(['SPREAD', 'HSICIG', 'DIVAMT'], 1)
df = df.drop(df.index[3364])
df.columns.values[0] = 'permno'

print(df.isnull().sum())

df['date'] = df.date.str[3:]

display(df.head())
```

```python
# Load in the Ratios
ratios = pd.read_csv('Data/Ratios.csv', sep=',', header=0)

ratios = ratios.drop(['sale_nwc', 'adate', 'qdate'], 1)

ratios.columns.values[1] = 'date'
ratios['date'] = ratios.date.str[3:]

display(ratios.head())
```

```python
# Merging the two dataframes
data = pd.merge(df, ratios, left_on=['date', 'permno'], right_on=['date', 'permno'])
display(data.head())
```
